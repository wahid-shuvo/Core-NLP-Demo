```python
from pycorenlp import StanfordCoreNLP
import json

nlp = StanfordCoreNLP('http://localhost:9000')

doc = "Ronaldo has moved from Real Madrid to Juventus. While messi still plays for Barcelona"
annot_doc = nlp.annotate(doc,
    properties={
        'annotators': 'ner, pos',
        'outputFormat': 'json',
        'timeout': 1000,
    })
```

```python
print(annot_doc)
```

```python
annot_doc = nlp.annotate(doc,
    properties={
        'annotators': 'sentiment',
        'outputFormat': 'text',
        'timeout': 1000,
    })
```

```python
print(annot_doc)
```
